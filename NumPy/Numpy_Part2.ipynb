```python
# ARRAY BOOLEAN INDEXING
# Boolean indexing means using True/False conditions to filter or select elements from an array.
# NumPy checks each element → makes a True/False array → returns only True positions
```

```python
arr = np.arange(1,21)
arr
```

```python
bool_values = arr > 10
bool_values
```

```python
arr = arr[bool_values] # reassignment
arr
# We can do it in one line 
# arr = arr[arr > 10]
# arr
```

```python
a = np.array([10, 20, 30, 40, 50])

# AND — both conditions must be True
a[(a > 15) & (a < 45)]    # [20 30 40]

# OR — either condition is True
a[(a < 15) | (a > 45)]    # [10 50]

# NOT — inverts the condition
a[~(a > 25)]              # [10 20]  (same as a <= 25)
```

```python
import numpy as np

# Valid examples
A = np.ones((8, 4, 6))
B = np.ones((1, 1, 6))
print((A + B).shape)   # (8, 4, 6) ✓

A = np.ones((8, 4, 6))
B = np.ones((8, 1, 6))
print((A + B).shape)   # (8, 4, 6) ✓

A = np.ones((4, 3))
B = np.ones((1, 3))
print((A + B).shape)   # (4, 3) ✓

A = np.ones((3, 1))
B = np.ones((1, 4))
print((A + B).shape)   # (3, 4) ✓

# Invalid examples — ye sab ValueError denge
try:
    A = np.ones((8, 4, 6))
    B = np.ones((7, 1, 6))
    print((A + B).shape)
except ValueError as e:
    print("Error:", e)   # Error ✗

try:
    A = np.ones((4, 3))
    B = np.ones((4,))
    print((A + B).shape)
except ValueError as e:
    print("Error:", e)   # Error ✗

try:
    A = np.ones((8, 1, 6))
    B = np.ones((7, 1, 5))
    print((A + B).shape)
except ValueError as e:
    print("Error:", e)   # Error ✗
```
