```python
# DEEP and SHALLOW COPY
# 1. SHALLOW COPY
# A shallow copy creates a new array object but does not create new data in memory  
# it still points to the same data as the original. So if you change the data, 
# the original gets affected too.

A = np.arange(1,10)
B = A.view()
B[0] = 15
print(A)
print(B)

# View creates a new array object but shares the underlying data buffer. So:
# Data changes → affect both
# Shape/dtype changes → independent

# 2. DEEP COPY
# A deep copy creates a new array object AND new data in memory  
# completely independent from the original. Changing it will never affect the original.

X = np.arange(11, 21)
Y = X.copy()
Y[0] = 55
print(X)
print(Y)
```

```python
# MATRIX MULTIPLICATION
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[7,8] , [3,4]])
print(mat1 * mat2 , end="\n\n") # it is just index multiplication
print(mat1 @ mat2 , end="\n\n") # it is matrix multplication
# We can also use dot function 
print(np.dot(mat1, mat2))
```

```python
#STACKING Of ARRAY  
arr = np.array([1,2,3,4])
arr2 = np.array([5,6,8,9])
print(np.vstack((arr , arr2))) # join vertically
print(np.hstack((arr , arr2))) # join horizontally
print(np.column_stack((arr, arr2))) #join vertically and convert into 2D matrix
print(np.stack((arr , arr2) , axis=0)) # 0 for column -- same as vstack
print(np.stack((arr , arr2) , axis=1)) # 1 for row -- same as column stack
```

```python
# On 2D Arrays
mat1 = np.array([[1,2] , [2,3] , [3,4]])
# Actually look like this
# [[1 2]
#  [2 3]
#  [3 4]]
mat2 = np.array([[2,3] , [4,5] , [9,0]])
print(np.vstack((mat1 , mat2)))
print(np.hstack((mat1 , mat2)))
print(np.column_stack((mat1 , mat2)))
```
