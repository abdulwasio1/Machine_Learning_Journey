```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
np.sum(s , axis=1) # column wise sum
```

```python
np.sum(s, axis=0) #row wise sum
```

```python
s[0:3 , 0:3]
s[0:3 ,3:6 ]
# 3 x 3 block sum check
for i in range(0,9,3):
    for j in range(0,9 , 3):
        sum = np.sum(s[i:i+3 , j:j+3])
        print(sum)
```

**General Questions**

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
data.shape
```

```python
#Find the average age of students.
print(np.mean(data[: , 0])) # first extract all the rows and then select column slicing
```

```python
#Extract Math marks of all students.
print(data[: , 1])
```

```python
#Find the highest Science mark.
print(data[:,2].max())
```

```python
#Get details of the student who scored more than 90 in Math.
#Here we get details of students also so we use "data[...]" outside bcz we have to extract the data
print(data[data[:,1] > 90])
# so it will print the details of student 2
```

```python
#Increase Math marks of all students by 5.
data[:,1] += 5
data
```

```python
#Find how many students are younger than 19.
print(data[data[:,0] < 19])
len(data[data[:,0] < 19])
```

```python
#Calculate the average marks in each subject (column-wise mean).
np.mean(data[: , 1:3] , axis= 0)
```

```python
#Get data of students who scored at least 80 in both subjects.
data[(data[:,1] > 80) & (data[:,2] > 80)] # make sure to use proper commas
```

```python
#Replace all Science marks < 75 with 0
data[:,2][data[:,2] < 75] = 0
data
#sab sa phle ham na wo columns extract kia phir use column ko select kr ka condition apply ki then usko zero ka equal kr dia 
```
