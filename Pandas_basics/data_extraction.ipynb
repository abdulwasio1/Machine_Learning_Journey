```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Countries.csv")
data.head()
```

**Q1. Which Country has highest population**

```python
# data.isna().sum()
data[data["population"] == data["population"].max()]["country"]
```

**Q2. What is the Capital of country having highest population**

```python
data[data["population"] == data["population"].max()]["capital_city"]
```

**Q3. Which country has least Population**

```python
data[data["population"] == data["population"].min()]["country"]
```

**Q4. What is the Capital of country having least population**

```python
data[data["population"] == data["population"].min()]["capital_city"]
```

**Q. Give me 5 countries with highest democartic score**

```python
data.nlargest(5 , "democracy_score")[["country" , "democracy_score"]]

#WE can use sorting also

value = data.sort_values(by ="democracy_score" , ascending= False) # it will not affect actual data unless i use 'inplace' or equals to orignal column
value["country"].head()
```

**Q. How many total regions are**

```python
# count = data["region"].count() --> It is wrong bcz it just give row count not count based on region
count = data["region"].value_counts().count() # value count tells us individual count based on each region, count will tell total region count
print(count)
```

**Q. How many countries lie in Eastern Euopriean region**

```python
count = data[data["region"] == "Eastern Europe"]["country"].count()
# Alternative Approach
count = data["region"].value_counts()["Eastern Europe"]
print(count)
```

**Q. Who is the political leader of 2nd highest populated country**

```python
data.nlargest(2 , "population")[["country" , "political_leader"]].iloc[1] #Index based selection bcz two largest values are extracted one is on 0th index and 2nd is on 1st 

#Alternative
data[data["population"] == data["population"].nlargest(2).iloc[1]][["political_leader"]]
```

**Q. How many countries whose political leaders are unknown**

```python
count = data[data["political_leader"].isna()]["country"].count()
print(count)
```

**Q. How many countries have Republic in their full name**

```python
# count = data[data["country_long"].]["country"].count()
# print(count)
count = 0
def count_names(text):
    global count
    if "Republic" in text:
        count += 1
data["country_long"].apply(count_names)
print(count)
```

```python
#Alternative and Better approach
count = data["country_long"].str.contains("Republic").sum()
print(count)
```

**Q. Which country in Africa region has highest population**

```python
value = data[data["continent"] == "Africa"]
value[value["population"] == value["population"].max()]["country"]
```
