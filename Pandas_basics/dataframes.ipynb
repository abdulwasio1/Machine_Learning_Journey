```python
#INDEXING AND SLICING
```

```python
data = {
    'Name' : ["Wasio" , "Nisar" , "Sherri" , "Sarmad" , "Hasaan"],
    'Age' : [19 , 20 , 19.5, 20 , 21],
    'City' : ["Gambat" , "Mirpur" , "Khairpur" , "Sukkur" , "Karachi"]
}
df2 = pd.DataFrame(data)
print(df2)
```

```python
#.loc - location - Label based indexing (row ,  columns) 
# sub_loc = df2.loc[1:3 , 2:3] # Error bcz we can not use indexes for both but we can use for rows (0-1 - by default generated for rows)
sub_loc = df2.loc[1:3 , ["Name" , "City"]]
sub_loc
#here 1:3 indicates rows indexes 1 to 3 (3 is included) and 2nd Shows which columns you want (Labels)
```

```python
#.iloc
# sub_iloc = df2.iloc[1:3 , ["Name" , "City"]] # error bcz iloc requires numeric indexers for both rows and columns
sub_iloc = df2.iloc[1:3 , 0:2] #Here last is excluded like 0 : 2, so 0 is included and 2 is excluded
sub_iloc
```

```python
'''
| Feature              | `loc`                  | `iloc`                       |
| -------------------- | ---------------------- | ---------------------------- |
| Access by            | Label (index name)     | Position (integer index)     |
| Uses                 | Index labels           | Row/column numbers           |
| End value in slicing | Included               | Excluded                     |

'''
```

```python
#WE CAN DO SELECTION BASED ON BOOLEAN VALUES
```

```python
values = df2['Age'] > 19.5
df3 = df2[values]
df3
```

```python
# WE CAN ALSO DO MODIFICATION on the basis of values - altough we can do it on the basis of index but for large data index count would be difficult
df2.loc[df2["Age"] == 20 , "City"] = "Nawabshah" # iska mtlb un records ki city nikalo jinki age 20 ha, or unki city ko ab "Nawabshah" kr lo
print(df2)
```

```python
#NOTE :  We can also use custom indexes
df = pd.DataFrame({
    "Name": ["Ali", "Sara", "Ahmed"],
    "Age": [20, 22, 21]
}, index=["a", "b", "c"])

print(df.loc['a' : 'c']) # Row Label = 'a'  to 'c' (included) & all columns
print(df.loc[['a' , 'b'] , "Name"]) # Row label = "a , b" & Column label = "Name"
print(df.iloc[0:3 , 0:2]) # Row Position = 0 to 3 (excluded) & column position = 0 to 2 (excluded)
print(df.iloc[1,0]) # Row position = 1 & Column position = 1
```
