```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("anime.csv")
data.head()
```

**Q. Make a Seperate Column of Episodes**

```python
data.loc[2 , "Title"] #.loc is label-based indexing

def extract_ep(text):
    start = text.find("(")
    end = text.find(")")

    if start != -1 and end != -1:
        result = text[start+1:end]
    return result

data["Episodes"] = data["Title"].apply(extract_ep)

```

```python
data["Episodes"] = data["Episodes"].str.replace(" eps" , "") # Top perform Numeric Operation
data
#STILL Our Episodes are String, now we have to conevert it int int
```

```python
data["Episodes"] = data["Episodes"].astype(int)
print(type(data.loc[2 , "Episodes"]))
```

**Q. Make a Seperate Column of TimeStamp**

```python
def extract_dates(text):
    start = text.find(")")

    if start != -1:
        result = text[start+1: start + 20]
    return result
data["Date & Time"] = data["Title"].apply(extract_dates)
```

**Q. Which Anime has highest Score**

```python
data[data["Score"].max() == data["Score"]]
```

**Q. Top 5 Highest Scores**

```python
data.nlargest(5 , "Score")[['Rank', 'Title', 'Episodes', 'Score']]
```

```python
data[data["Episodes"] == data["Episodes"].max()]
```
