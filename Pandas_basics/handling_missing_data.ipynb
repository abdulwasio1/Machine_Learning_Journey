```python
import numpy as np
import pandas as pd
```

```python
data = {
    "A": [1, 2, 3, 4, 5, 6],
    "B": [10, np.nan, 30, 40, np.nan, 60],
    "C": [100, 200, 300, np.nan, 500, 600],
    "D": [7, 8, 9, 10, np.nan, 12],
    "E": [15, np.nan, 17, np.nan, 19, np.nan]
}
df = pd.DataFrame(data) # by default it took datatype as float except column A bcz of no any null value (it works column basis)
df
```

**CHECKING NULL VALUES (Works Column basis)**

```python
df.isna()
```

```python
df.isna().sum() # Count null values in each column
```

```python
df.isna().any() #tells kon konsa column ma null values han
```

**Removing Missing Data (Works ROW basis)**

```python
df.dropna() # ya har us row ko remove kr dega jis ma null values han
```

```python
# We can set a threshold or limit to check n number of not null values like if i set 
# theshold to 4, them it must force you to extract those rows which has atleast  4 non null values
df.dropna(thresh=4)
```

**FILLING THE MISSING DATA (Works Column wise)**

```python
df.fillna(0) # Fill all the NaN values with zero
# We can use custom values in each column
df.fillna({'A' : 10 , 'B' : 20 , 'C' : 30 , 'D':50 , 'E' : 60}) # works columns wise
df.fillna(df.mean()) 
df.fillna(df.median())
#Calculate mean of each column and Replace every NaN in a column with that column's mean.
```

```python
# There is Also Backward Fill (next valid value) and Forward Fill (previous valid value)
print(df)
print(df.bfill())
print(df.ffill())
```

**For removing - dropna(),**
