```python
data = ('Name , Age , City\n'
       'Wasio , 19 , Gambat\n' 
       'Nisar , 20 , Mirpur\n'
       'Hasaan , 21 , Khairpur\n')
# Above is the actual format of CSV file
type(data)
#now i have to convert this str into file formarted data basically term used for StringIO -> in memory file format object 
df = pd.read_csv(StringIO(data))
df
```

**Reading Specific columns**

```python
df = pd.read_csv("products.csv" , usecols= ['Name', 'Brand' , 'Price' , 'Size'])
# We can convert this sub dataset into csv file back using 'to_csv' function
df.to_csv('fileName.csv' , index= False) # False means it does not store the indexes now
df.head()
```

```python
numbers = ('a , b , c , d\n'
          '1,2,3,4\n'
          '3,4,5,6\n'
          '8,9,0')
df = pd.read_csv(StringIO(numbers))
df
print(df.dtypes) # last column has null value so Pandas take by  default as NaN (is float) so entire column become float 
#BUT we can specify the datatype intially at the time of file reading

# df2 = pd.read_csv(StringIO(numbers), dtype= 'int')  # Got error bcz it contain NaN value
df2 = pd.read_csv(StringIO(numbers), dtype= 'float')
df2.dtypes
df2.info() #clearly shows column 4 contain 2 not null values means 1 null values is present also
df2 # all are converted into Float values
```

```python
 # We can also see null values using 'is_null' function
print(df2.isnull())
print(df2.isnull().sum())
```

***Custom DataTypes***

```python
numbers = ('a,b,c,d\n'
          '1,2,3,4\n'
          '3,4,5,6\n'
          '8,9,0')
df = pd.read_csv(StringIO(numbers) , dtype= {'a': int , 'b' : object , 'c' : int})
df.dtypes
df.info()
```

***CUSTOM INDEXES***

```python
data = ('Index,Name,Age,City\n'
       '2,Wasio , 19 , Gambat\n' 
       '3,Nisar , 20 , Mirpur\n'
       '6,Hasaan , 21 , Khairpur\n')
#Here we want to use 'Index' column as Index not by default index (0-n)
df = pd.read_csv(StringIO(data) , index_col=0) # 0 is number of column either 1st , 2nd or zeroth
df2 = pd.read_csv(StringIO(data) , usecols=['Name' , 'Age'] , index_col=0) #here index is 0th column
#NOTE : Now Pandas looks at the first remaining column (position 0), which is Name, and makes it the index.
print(df2)

# If you wanted Index as the index Then you must include it in usecols:
df2 = pd.read_csv(StringIO(data), usecols=['Index', 'Name', 'Age'], index_col=0)
print(df2)

# index_col=0 means:
# Use the first column after all column selections (usecols) have been applied, not necessarily the first column in the original CSV.
```
