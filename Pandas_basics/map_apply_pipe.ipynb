```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("sales.csv" , parse_dates=["Date"] , dtype={'PaymentMethod' : "category"})
data.info()
data.head()
```

# MAP FUNCTION

```python
# Map is used on Pandas series so we can use a single column of Dataframe as a series
data.groupby("PaymentMethod").size()
data["Payment Type"] = data["PaymentMethod"].map({
    "Bank Transfer" : "Physical",
    "COD" : "Physical",
    "Credit Card" : "Digital",
    "Debit Card" : "Digital",
    "PayPal" : "Digital"
})
# It creates new column and fill values according to above given condition like if PayementMethod id Bank Transfer/COD, fill that rows with Physical Payment Type
data.info()
data.head()
data.groupby("Payment Type").size()
```

# APPLY Operation

```python
def rating_category(data):
    # print(type(data)) -> Always Series
    if data["Rating"] == 5:
        return "Best"
    elif data["Rating"] == 4:
        return "Good"
    elif data["Rating"] == 3:
        return "Average"
    elif data["Rating"] == 2:
        return "Below Average"
    else : return "Bad"
data["Rating Category"] = data.apply(rating_category , axis=1 ) 
# Note if i am apply on sigle column/row then i don't have to mention axis other for entire data i have to mention the axis (1 always other wise error)
# WHY AXIS is 1? : bcz data that i am passing as arrgument is a single row and we are accessing rating column from that row -> row["Rating"]
# Here each row is passing as a series 
'''
OrderID   Rating   GrandTotal
0     ORD-001      5        250.0
1     ORD-002      3        120.0
2     ORD-003      4        390.0

axis=0 → function receives ['ORD-001', 'ORD-002', 'ORD-003']  ← entire column
axis=1 → function receives ['ORD-001', 5, 250.0]              ← entire row
'''
print(data[["Rating" , "Rating Category"]].tail() , end="\n\n")
print(data["Rating Category"].value_counts() , end="\n\n") # Like group by
print(data.groupby("Rating Category")["Rating Category"].count())
```

# PIPE FUNCTION
**pipe = pass the DataFrame into a function, get it back, pass to next function**

**Water → Pipe 1 (Filter) → Pipe 2 (Add Minerals) → Pipe 3 (Cool Down) → Final Water**

**Same in pandas: DataFrame → pipe(add_profit) → pipe(add_net_revenue) → Final DataFrame**

```python
def add_profit(data):
    data['Profit'] = data['GrandTotal'] * 0.20   # 20% profit margin
    return data

def add_net_revenue(data):
    data['NetRevenue'] = data['GrandTotal'] - data['Profit']
    return data
data = (
    data
    .pipe(add_profit)
    .pipe(add_net_revenue)
)
data.head()
```
