**Merging 2 Dataframes**

```python
import numpy as np
import pandas as pd
```

```python
names = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha", "Bilal"],
    "Department": ["IT", "HR", "Finance", "IT", "Marketing"]
})
salaries = pd.DataFrame({
    "ID": [1, 2, 3, 7, 8],
    "Salary": [50000, 60000, 55000, 70000, 65000],
    "Experience": [1, 3, 2, 5, 4]
})
```

```python
names
salaries
```

**MERGE(Work same like JOIN in MySQL)**

```python
pd.merge(names , salaries) #Automatically find comman values (ID) to merge but we can also set manually
```

```python
pd.merge(names , salaries , on='ID' , how='inner')
# As in MySQL there is inner, outer , left and right join - but default is inner join but we can change using 'how=' parameter
pd.merge(names ,salaries , on='ID' , how='outer')
pd.merge(names ,salaries , on='ID' , how='left')
pd.merge(names ,salaries , on='ID' , how='right')
```

**Concatination of 2 Dataframes**

```python
names = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha", "Bilal"],
    "Department": ["IT", "HR", "Finance", "IT", "Marketing"]
})
salaries = pd.DataFrame({
    "ID": [1, 2, 3, 7, 8],
    "Salary": [50000, 60000, 55000, 70000, 65000],
    "Experience": [1, 3, 2, 5, 4]
})
```

```python
# pd.concat(names , salaries) # Error bcz concat takes one argument but we have two values so we pass list instead of direct values
pd.concat([names , salaries]) # By default it works column basis - axix 0
# But we can change axis 1 (Row)
pd.concat([names , salaries] , axis=1) #Ro w basis
```

**Joining of two dataframes**

```python
A = pd.DataFrame({
    "key": [1, 2, 3, 4],
    "x": [100, 200, 300, 400]
})

b = pd.DataFrame({
    "key": [3, 4, 5, 6],
    "y": [3000, 4000, 5000, 6000]
})
```

```python
# a.join(b) gets error bcz "JOIN BY DEFAULT WORKS ON INDEX"
# UPDATED DATAFRAMES
a = pd.DataFrame({
    "x": [100, 200, 300, 400]
} , index= [1, 2, 3, 4])

b = pd.DataFrame({
    "y": [3000, 4000, 5000, 6000] 
} , index=[3, 4, 5, 6]) 
print(a.join(b)) # BASICALLY IT IS INNER JOIN
# See the clear difference b/w merge and join. Merge only keeps comman rows values but joins keeps both (with NaN values)
# We Can change by default comparssion values
print(a.join(b , how='outer'))
```
