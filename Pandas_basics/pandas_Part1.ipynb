```python
import pandas as pd
```

```python
#A Series in Pandas is a one-dimensional labeled array - we can make our own custom labels to determine values
s = pd.Series(["Wasio",20,20,30,12.34,55] , index=[1,2,3,1,4 ,"He"])
#if i do not mention labels then it will be by default - 0 , 1 , 2 , 3 , ...
```

```python
print(s[1])
print(s[2])
print(s["He"])
print(s[4])
#Can use Multiple datatypes
```

```python
print(s.value_counts())
```

```python
print(s.index)
print(s.values)
print(s.dtype)
```

```python
print(s.head)
print(s.tail)
#Print some itital and ending data - majorily works for large datasets
```
