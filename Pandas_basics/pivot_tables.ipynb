```python
import numpy as np
import pandas as pd
```

```python
#SAMPLE DATA
data = {
    'Date':     pd.date_range('2024-01-01', periods=30),
    'City':     np.random.choice(['Karachi', 'Lahore', 'Islamabad', 'Quetta', 'Peshawar'], 30),
    'Product':  np.random.choice(['Tea', 'Rice', 'Sugar', 'Flour', 'Oil'], 30),
    'Category': np.random.choice(['Grocery', 'Beverage', 'Staple'], 30),
    'Sales':    np.random.randint(100, 1000, 30),
    'Units':    np.random.randint(1, 50, 30),
    'Rep':      np.random.choice(['Ali', 'Sara', 'John', 'Nadia', 'Usman'], 30),
    'Returned': np.random.choice([True, False], 30, p=[0.1, 0.9]),
}

df = pd.DataFrame(data)
df
```

```python
'''
why we use pivot tables in pandas?
Pivot tables are used to summarize and reorganize large data into a meaningful format — 
instead of staring at hundreds of raw rows, you get a clean summary in seconds.
'''
pf = pd.pivot_table(df , index="City" , columns="Product" , values="Sales") # You should have to mention these values to get meaningful data
pf
# data -> Dataframe 'df'
# index -> City 
# columns -> Product
# values -> Sales (Use Numeric values bcz of aggregation)
#Now some Sales values are NaN it is because 'Some Cities has not any product Sale like Karachi has no any sale of Rice - also in actual data'
#NOTE : Below values is not actual value, it is the mean of all each column like -> Isb ma Rice ki Sales ha 960 and 316 so 638 is the mean of both 
# MEAN is set as by default, but we can change it 
pf = pd.pivot_table(df , index="City" , columns="Product" , values="Sales" , aggfunc='sum')
pf

pf = pd.pivot_table(df , index="City" , columns="Product" , values= ["Sales" , "Units"] , aggfunc='sum') #Can use multiple values (Numeric, nhi to error)
pf
```
