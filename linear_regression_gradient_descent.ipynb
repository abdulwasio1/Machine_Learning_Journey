```python
import numpy as np
```

```python
size = np.array([500,800,1000,1200,1500])
price = np.array([100,150,190,230,280])
size_norm = size / 1000 
price_norm = price / 100
num = int(input("Enter the value you want to predict : "))
learning_rate = 0.1
epoch = 500
m = 0
b = 0
n = len(size_norm)
for i in range(epoch):
    prediction = m * size_norm + b
    dm = (-2/n) * np.sum(size_norm * (price_norm - prediction))
    db = (-2/n) * np.sum(price_norm - prediction)
    m = m - learning_rate * dm
    b = b - learning_rate * db
final_prediction = m * size_norm + b
final_cost = np.mean((price_norm - final_prediction) ** 2)
print("The Best Value of m under 500 Epochs is",m)
print("The Best Value of b under 500 Epochs is",b)
print("Final Cost (MSE) after 500 epochs:",final_cost)
result_norm = m * (num/1000) + b
result = result_norm * 100
print("The Price of", num, "sq.ft house is $", result)
    
```
